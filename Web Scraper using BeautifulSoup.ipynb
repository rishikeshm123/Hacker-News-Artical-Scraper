<h4><font color = 'Darkblue'> Importing Libraries

```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

<h4><font color = 'Darkblue'>  Using Loop to get links of first 10 hacker news pages

```python
## fUNCTION TO GET lINKS OF FIRST 10 PAGES
root_url = "https://news.ycombinator.com"
def Get_Links(root_url):
    link = []                                # empty list to store links
    for i in range(0,11):                    #iterating through first 10 pages by navigating through subpages using root url
        l = root_url + "/news?p={}".format(i)
        link.append(str(l))
       
    return link  
```

```python
page_links = Get_Links(root_url)             #Links of first 10 pages
page_links
```

<h4><font color = 'Darkblue'>  Using Beautiful Soup to get News Title, News Links, and Points given by people.

```python
##fUNCTION TO GET TITLE AND TITLE LINKS
def Complete_Title_Links(page_links):
    complete_links = []
    for li in page_links:
        res = requests.get(li)
        soup = BeautifulSoup(res.text, "html.parser")
        title_links = soup.select('.titleline > a')
        complete_links = complete_links + title_links
    return complete_links 
```

```python
cc = Complete_Title_Links(page_links)       #HTML elements including Title and Links . cc = Title + Link (href)
cc
```

```python
##fUNCTION TO GET Points
def Get_Subtext(page_links):      
    subt = []
    for li in page_links:
        res = requests.get(li)
        soup = BeautifulSoup(res.text, "html.parser")
        su = soup.select('.subtext')
        subt = subt + su
    return subt 
```

```python
subtext = Get_Subtext(page_links)             #HTML elements including Points , Subtext = Points
subtext
```

<h4><font color = 'Darkblue'>  Functions to clean up text and create a dataframe of News Titles, News Links and Points

```python
def create_custom_hn(cc, subtext):  
    hn = []
    for idx, item in enumerate(cc):
        title = str(item.getText())
        linkk = str(item.get('href', None))
        points = int(subtext[idx].getText().split()[0])
        hn.append({"Title": title, "Link": linkk,'Points': points})
            
    return pd.DataFrame(hn)
```

```python
df = create_custom_hn(cc, subtext)  #dataframe including all news titles ,links and points 
df
```

```python
df = df[df['Points'] >= 100].sort_values(by = 'Points', ascending = False)  #filtering out news titles with over 100 points
df
```

```python
# Replacing all ASK HN articles and appending the root url to them
# root url + "item?id"
df['Link'].loc[df["Link"].str.startswith('item?id', na=False)] = df['Link'].loc[df["Link"].str.startswith('item?id', na=False)].apply(lambda x: "{}{}".format("https://news.ycombinator.com/", x))
```

```python
#Final Clean Dataframe 
df 
```
